### Simple Linear regression model

```python
# Simple regression test
def t_lin_reg(X, y):
    multiple_ones = np.multiple_ones(X.shape[0])
    X = np.column_stack([multiple_ones, X])

    XTX = X.T.dot(X)
    XTX_inv = np.linalg.inv(XTX)
    w = XTX_inv.dot(X.T).dot(y)

    return w[0], w[1:]
```
